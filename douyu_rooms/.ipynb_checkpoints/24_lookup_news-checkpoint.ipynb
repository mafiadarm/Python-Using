```python
import re
import requests
import json
import os
from multiprocessing.dummy import Pool
```

# 开始

```python
def get_class():
    print('-------------- 读取大类 --------------')
    response = requests.get('https://www.douyu.com/directory/all')
    rex = r'href="(/directory\?shortName=.*?)"'
    return re.findall(rex, response.text)
```

```python
def directory(dd):
    response = requests.get(f'https://www.douyu.com{dd}')
    rex = '(\{"cate2Name".*?\})'
    class_dict = [json.loads(i) for i in re.findall(rex, response.text)]
    class_url = ['https://www.douyu.com' + i.get('cate2Url') for i in class_dict]
    for url in class_url:
        yield url
```

```python
def plate(url):
    response = requests.get(url)
    rex = r'(\{"authInfo".*?\},.*?"vipId".*?\}),'
    result = re.findall(rex, response.text)

    for i in result:
        try:
            tail = json.loads(i).get('url')
            room_url = 'https://www.douyu.com' + tail
            yield room_url
        except:
            pass
```

```python
def save(content):
    with open('douyu_topic.txt', 'a+') as ww:
        ww.write(f'{content}\n')

def save_lines(file_name, content):
    with open(file_name, 'w') as ww:
        ww.writelines(content)

def load(file_name):
    with open(file_name, 'r+') as rr:
        content = rr.readlines()
    return [i.replace("\n", "") for i in content]
```

```python
def auth_data(room_id):
    response = requests.get(room_id)
    rex = r'window.zhtName="(.*?)",window.room_ids=\[(.*?)\],'
    result = re.findall(rex, response.text)
    if result:
        result = result[0]
        print(f'{result[0]} 正在直播...')
        save(result[0])
        return True
```

```python
def run_pro(tail):
    # print(f'--start {tail}--')
    for room_id in plate(tail):
        # print(f'test room {room_id}')
        if auth_data(room_id):
            print(f'退出 {tail}')
            return
```

```python
def compare():
    new = load('douyu_topic.txt')
    old = load('douyu_topic_old.txt')
    ing = set(new) ^ set(old)
    
        
    if not ing:
        print('no room close')
        print('no room open')
        return
    
    close = set(old) & ing
    if close:
        for c in close:
            print(f'close: {"https://douyu.com/" + c}')
    else:
        print('no room close')
              
    open_ = set(new) & ing
    if open_:
        for o in open_:
            print(f'open: {"https://douyu.com/" + o}')
    else:
        print('no room open')

    _n = '\n'.join(new)
    save_lines('douyu_topic_old.txt', _n)
    save_lines('douyu_topic.txt', '')
```

```python
if not os.path.exists('douyu_topic_old.txt'):
    with open('douyu_topic_old.txt', 'w') as ww_init:
        ww_init.write('')
        
if not os.path.exists('douyu_topic.txt'):
    with open('douyu_topic.txt', 'w') as ww_init:
        ww_init.write('')        

pool = Pool(16)    
    
for sn in get_class():
    for tail in directory(sn):
        # print(f'--------------游戏直播 {tail} --------------')
        pool.apply_async(run_pro, (tail,))

pool.close()
pool.join()
```

```python
compare()
```
